```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Hela-CalcienAM-stats.csv')
a = data[data['Status'] == 'Apoptotic']
#d.loc[:,'Status'] = 'Apoptotic'
div = data[data['Status'] == 'Dividing']
#div.loc[:,'Status'] = 'Dividing'
nd = data[data['Status'] == 'Non-dividing']
#h.loc[:, 'Status'] = 'Healthy'
dead = data[data['Status'] == 'Dead']
data = nd.append(div)
data = data.append(a)
data = data.append(dead)
```

```python
data = pd.read_csv('Hela-CalcienAM-stats.csv')
```

```python
plt.figure(figsize = (50,50))
sns.lmplot(x="Area", y="Mean Intensity", hue="Status",
               truncate=True, size=4, data=data)
plt.show()
```

```python
ave = pd.DataFrame(data.groupby('Status').mean()['Mean Intensity'])#.reset_index()
ave
```

```python
std = pd.DataFrame(data.groupby('Status').std()['Mean Intensity'])#.reset_index()
std
```

```python
count = pd.DataFrame(data.groupby('Status').count()['Mean Intensity'])#.reset_index()
count
```

```python
df = ave.join(std,lsuffix = '_ave', rsuffix = '_std')
df = df.join(count, rsuffix = '_count')
df
```

```python
df.columns = ['Ave','Std','Counts']
df['Error'] = df['Std']/((df['Counts'])**(1/2))
df = df.iloc[::-1]
df
```

```python
#x = ['Apoptotic (N=73)','Dead (N=74)', 'Dividing (N=40)','Non-dividing (N=79)'][::-1]
axis = ['Non-dividing (N=79)', 'Dividing (N=40)',  'Apoptotic (N=73)','Dead (N=74)']
z = ['a','b','d','c']
```

```python
average = df['Ave'].tolist()
errors = df['Error'].tolist()
#ave = [average[0],average[1],average[3],average[2]]
#err = [error[0],error[1],error[3],error[2]]
```

```python
num = list(range(len(z)))
num
```

```python
plt.figure(figsize = (10,7))
plt.bar(z,average, yerr = errors)
plt.ylabel('Average MFI',fontsize=18)
plt.ylim(0,45000)
plt.yticks(fontsize = 12)
#plt.xticks(('Non-dividing (N=79)', 'Dividing (N=40)', 'Dead (N=74)', 'Apoptotic (N=73)'),fontsize = 14)
plt.xticks(num,axis,fontsize = 14)

x0, x1, x2, x3 = 0, 1, 2, 3  
y, h, col = df['Ave'].max() + 2000, 400, 'k'
plt.plot([x0+0.05, x0+0.05, x1-0.05, x1-0.05], [y, y+h, y+h, y], lw=1.5, c=col)
plt.plot([x1+0.05, x1+0.05, x2-0.05, x2-0.05], [y, y+h, y+h, y], lw=1.5, c=col)
plt.plot([x2+0.05, x2+0.05, x3-0.05, x3-0.05], [y, y+h, y+h, y], lw=1.5, c=col)
plt.plot([x0+0.05, x0+0.05, x2-0.05, x2-0.05], [y+3000, y+h+3000, y+h+3000, y+3000], lw=1.5, c=col)
plt.plot([x1+0.05, x1+0.05, x3-0.05, x3-0.05], [y+6000, y+h+6000, y+h+6000, y+6000], lw=1.5, c=col)
plt.plot([x0+0.05, x0+0.05, x3-0.05, x3-0.05], [y+9000, y+h+9000, y+h+9000, y+9000], lw=1.5, c=col)



plt.text((x0+x1)*.5, y+h, "*", ha='center', va='bottom', color=col,fontsize = 14)
plt.text((x1+x2)*.5, y+h, "ns", ha='center', va='bottom', color=col,fontsize = 14)
plt.text((x2+x3)*.5, y+h, "*", ha='center', va='bottom', color=col,fontsize = 14)
plt.text((x0+x2)*.5, y+h+3000, "*", ha='center', va='bottom', color=col,fontsize = 14)
plt.text((x1+x3)*.5, y+h+6000, "*", ha='center', va='bottom', color=col,fontsize = 14)
plt.text((x0+x3)*.5, y+h+9000, "*", ha='center', va='bottom', color=col,fontsize = 14)



plt.show()

```
