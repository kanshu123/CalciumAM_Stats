```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('231-CalcienAM-stats.csv')
```

```python
data = data.drop('Unnamed: 0',axis = 1)
```

```python
data['Aspect Ratio'] = data['Major']/data['Minor']
data.describe()
```

```python
data['nor_intensity'] = data['Mean Intensity']*10000
data.head()
```

```python
sns.jointplot("Area", "nor_intensity", data=data,xlim=(0,1),ylim=(0,600000),kind="reg", color="k", size=6)
#plt.ylabel('MFI',fontsize = 20)
#plt.yticks(fontsize = 14)
#plt.xlabel('Area',fontsize = 20)
#plt.xticks(fontsize = 14)
plt.show()
```

```python
sns.jointplot("Area", "Aspect Ratio", data=data, kind="reg", xlim=(0, 1), ylim=(0, 5),color="k", size=4,stat_func = None)
```

```python
sns.jointplot("Area", "Mean", data=data,  kind="kde", space=0, color="r")
plt.show()
```

```python
sns.jointplot("Aspect Ratio", "Mean Intensity", data=data, kind="reg",
              xlim=(0, 20), ylim=(0, 55), color="r", size=4)#,stat_func = None)
```

```python
sns.jointplot("Aspect Ratio", "Mean", data=data,  kind="kde", space=0, color="r")
plt.show()
```

```python
sns.jointplot("Circ.", "Mean Intensity", data=data, kind="reg",
              xlim=(0, 1), ylim=(0, 55), color="k", size=4,stat_func = None ) 
plt.ylabel('MFI',fontsize = 20)
plt.yticks(fontsize = 14)
plt.xlabel('Circularity',fontsize = 20)
plt.xticks(fontsize = 14)
plt.show()
```

```python
sns.jointplot("Round", "Mean", data=data, kind="reg",
              xlim=(0, 1), ylim=(0, 55), color="r", size=7)
```
