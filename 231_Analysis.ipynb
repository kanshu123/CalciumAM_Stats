```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('231-CalcienAM-stats.csv')
```

```python
data = data.drop('Unnamed: 0',axis = 1)
```

```python
data['Aspect Ratio'] = data['Major']/data['Minor']
data.describe()
```

```python
data.head(10)
```

```python
sns.jointplot("Area", "Mean Intensity", data=data, kind="reg",
              xlim=(0, 3), ylim=(0, 55), color="k", size=4,stat_func = None)
plt.show()
```

```python
sns.jointplot("Area", "Mean", data=data,  kind="kde", space=0, color="r")
plt.show()
```

```python
sns.jointplot("Aspect Ratio", "Mean", data=data, kind="reg",
              xlim=(0, 20), ylim=(0, 55), color="r", size=4)#,stat_func = None)
```

```python
sns.jointplot("Aspect Ratio", "Mean", data=data,  kind="kde", space=0, color="r")
plt.show()
```

```python
sns.jointplot("Circ.", "Mean Intensity", data=data, kind="reg",
              xlim=(0, 1), ylim=(0, 55), color="k", size=4,stat_func = None ) 
```

```python
sns.jointplot("Round", "Mean", data=data, kind="reg",
              xlim=(0, 1), ylim=(0, 55), color="r", size=7)
```
