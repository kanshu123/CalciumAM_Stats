```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('231-CalcienAM-stats.csv')
```

```python
data = data.drop('Unnamed: 0',axis = 1)
```

```python
data['Aspect Ratio'] = data['Major']/data['Minor']
data.describe()
```

```python
data.head(10)
```

```python
sns.jointplot("Area", "Mean", data=data, kind="reg",
              xlim=(0, 3), ylim=(0, 55), color="r", size=7)
plt.show()
```

```python
sns.jointplot("Aspect Ratio", "Mean", data=data, kind="reg",
              xlim=(0, 20), ylim=(0, 55), color="r", size=7)
```

```python
sns.jointplot("Circ.", "Mean", data=data, kind="reg",
              xlim=(0, 1), ylim=(0, 55), color="r", size=7)
```

```python
sns.jointplot("Round", "Mean", data=data, kind="reg",
              xlim=(0, 1), ylim=(0, 55), color="r", size=7)
```
